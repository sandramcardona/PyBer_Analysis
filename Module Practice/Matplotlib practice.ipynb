```python
#scatter plot using plt.scatter function - 5.1.7
#Skill drill 5.1.7
%matplotlib inline
# Import dependencies.
import matplotlib.pyplot as plt 
# Set the x-axis to a list of strings for each month.
x_axis = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
# Create the plot.
plt.scatter( y_axis, x_axis, c="red", label='Chicago')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 50.
plt.xlim(0, 50)
#invert axis -
plt.gca().invert_yaxis()
#plt.gca().invert_xaxis()
# Create a title.Invert the y-axis so that January is at the top
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#Create bubble chart - 5.1.7
plt.scatter(x_axis, y_axis)
```

```python
#Increase the size of circles
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Make the bubbles 3 times larger for y axis
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
#Then we can use the new y-axis list for the s parameter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#new_list = [expression for item in list if conditional]. 
#Use this inside the plt.scatter() function instead of the for loop below
#y_axis_larger = []
#for data in y_axis:
#.   y_axis_larger.append(data*3)
#
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Skill drill 5.1.7 - bubble chart

# Create the plot.
plt.scatter(x_axis, y_axis, color="skyblue", label='Boston',alpha=0.2, )
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#increase bubble size by 2
plt.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis])
#invert axis -
plt.gca().invert_yaxis()
# Create a title.Invert the y-axis so that January is at the top
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt() - skill drill 5.1.5
#how to add edgecolor black
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], alpha=0.2, edgecolor="black")
#change color and label
ax.scatter(y_axis, x_axis, color="skyblue", label='Boston', linewidth=2)
# Add the legend.
ax.legend()
# Create labels for the x and y axes.
ax.set(title = 'PyBer Fare by Month',
    xlabel = 'Fare($)',
    ylabel = 'Date')
plt.gca().invert_yaxis()
```
